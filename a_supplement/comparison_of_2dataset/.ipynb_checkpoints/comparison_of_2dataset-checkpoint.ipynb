```python
# file_path = os.path.join("Train_data", "supervised_data.csv")
file_path = os.path.join("Train_data", "updated_data_filtered.csv")
old_data = pd.read_csv(file_path)
old_data = old_data[old_data["high_level_substr"] != "capsule polysaccharide degradation"]
old_data
```

```python
# Extract the numeric part of the 'id' column
old_data['numeric_part'] = old_data['PUL ID'].str.extract('(\d+)', expand=False).astype(int)

# Sort the DataFrame based on the numeric part
old_data_sort = old_data.sort_values(by='numeric_part').reset_index(drop=True)
```

```python
file_path = os.path.join("Train_data", "supervised_data.csv")
# file_path = os.path.join("Train_data", "updated_data_filtered.csv")
data = pd.read_csv(file_path)
data
```

```python
data["high_level_substr"].value_counts()
```

```python
# Extract the numeric part of the 'id' column
data['numeric_part'] = data['PUL ID'].str.extract('(\d+)', expand=False).astype(int)

# Sort the DataFrame based on the numeric part
data_sort = data.sort_values(by='numeric_part').reset_index(drop=True)
```

```python
data_sort
```
